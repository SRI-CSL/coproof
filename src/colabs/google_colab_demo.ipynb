### &#129504; CoProver Simple T5 Model

```python
try:
    from simplet5 import SimpleT5
except ImportError:
    !pip install --upgrade simplet5
```

```python
IS_LOCAL_FILE = False
```

```python
DATA_DICT = {
    'tags_input_pred_pairs.tsv': '1J9xAcdlo8NXRlbz3SQh_GvbVsk90a4jK',
    'curr_best.zip': '1aOLRXmQACIimIGaqP3c-N58EaujSciaD'
    }
```

```python
def path_exists(input_path: str) -> bool:
    return os.path.exists(input_path)
```

```python
def check_file_status(input_path: str):
    if path_exists(input_path=input_path):
        print(f"- File {input_path.split('/')[-1]} exists locally at {input_path}!")
    else:
        if IS_LOCAL_FILE:
            print("- Something went wrong with the download. Please try again!")
        else:
            print(f"- IS_LOCAL_FILE is set to {IS_LOCAL_FILE}. The file is accessed via a public GitHub link!")
```

```python
def download_data_from_google_drive(google_file_id: str, output_file_name: str, quiet_download: bool) -> str:
    file_path = f'./{output_file_name}'

    if not os.path.exists(file_path):
        gdown.download(id=google_file_id, output=output_file_name, quiet=quiet_download)
    else:
        print(f"{output_file_name} already exists!")
        
    return file_path
```

```python
proof_file = download_data_from_google_drive(
    google_file_id=DATA_DICT['tags_input_pred_pairs.tsv'], 
    output_file_name='tags_input_pred_pairs.tsv', 
    quiet_download=False)
```

```python
check_file_status(input_path=proof_file)
```

```python
model_zip_file = download_data_from_google_drive(
    google_file_id=DATA_DICT['curr_best.zip'], 
    output_file_name='curr_best.zip', 
    quiet_download=False)
```

```python
check_file_status(input_path=model_zip_file)
```

```python
# read dataset into dataframe
full_df = pd.read_csv(f'./{proof_file}', sep="\t", header=None, names=['source_text', 'target_text'])
```

```python
full_df.info()
```

```python
full_df.describe()
```

```python
full_df.drop(full_df[(full_df.target_text == 'assert') | (full_df.target_text == 'tcc') | (full_df.target_text.str.endswith('-tcc'))].index, inplace=True)
```

```python
full_df.shape
```

```python
# Command prefixes are always expected
CMD1_PREFIX = "command1: "
```

```python
full_df['source_text'] = CMD1_PREFIX + full_df['source_text']
```

```python
# visualize
tmp_target_1 = full_df.groupby(full_df.target_text).size().sort_values()
tmp_target_1.plot.barh(title="Target Command Plots (after removing assert, tcc, and *-tcc)", xlabel="Commands", ylabel="Number of records", log=True, figsize=(15, 12))
```

```python
# Load a pre-trained T5 model
model = SimpleT5()
```

```python
model.from_pretrained(model_type="t5", model_name="t5-base")
```

```python
# unzip
import zipfile
with zipfile.ZipFile(f'./{model_zip_file}', 'r') as zip_ref:
    zip_ref.extractall('./curr_best')
```

```python
trained_model_path = './curr_best/curr_best'
```

```python
model.load_model("t5", trained_model_path, use_gpu=False)
```

```python
input_sequence = "<ANT> <CONS> s-formula forall ['variable'] ['variable'] apply constant type-actual apply constant type-actual type-actual apply constant ['variable'] ['variable'] apply constant apply constant type-actual type-actual ['variable'] apply constant type-actual type-actual ['variable'] <HID>"

# skolem!
print(model.predict(CMD1_PREFIX + input_sequence)[0])
```

```python
for idx, row in full_df.head(5).iterrows():
    input_sequence = row.source_text
    y_true = row.target_text
    y_pred = model.predict(input_sequence)[0]

    print(f"- Input sequence: {input_sequence}\n- True label: {y_true}\n- Predicted label: {y_pred}\n-------")
```
